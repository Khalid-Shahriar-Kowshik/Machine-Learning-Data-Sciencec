# Logistic Regression
Logistic regression is a supervised machine learning algorithm that accomplishes binary classification tasks by predicting the probability of an outcome, event, or observation. The model delivers a binary or dichotomous outcome limited to two possible outcomes: yes/no, 0/1, or true/false.

Logical regression analyzes the relationship between one or more independent variables and classifies data into discrete classes. It is extensively used in predictive modeling, where the model estimates the mathematical probability of whether an instance belongs to a specific category or not.

For example, 0 – represents a negative class; 1 – represents a positive class. Logistic regression is commonly used in binary classification problems where the outcome variable reveals either of the two categories (0 and 1

Some examples of such classifications and instances where the binary response is expected or implied are:

1. Determine the probability of heart attac.

2. Possibility of enrolling into a univeres.

3. Identifying spam 
4. Weather prediction
5. Classificationc value.

## Extracting data or importing data

```python
# importing libraries and collecting the data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline

data = pd.read_csv('Heart.csv')
```

```python
data.head(10)
```

```python
data.drop("Unnamed: 0",axis=1)
```

```python
data["ChestPain"]=data["ChestPain"].astype('category')
data["ChestPain"]=data["ChestPain"].cat.codes
data["Thal"]=data["Thal"].astype('category')
data["Thal"]=data["Thal"].cat.codes
data["AHD"]=data["AHD"].astype('category')
data["AHD"]=data["AHD"].cat.codes

data
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
data
```

```python
X= data.drop(columns='AHD')
X
```

```python
Y= data['AHD']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(X.shape)
print(Y.shape)
```

```python
# need to convert X and Y to arrays so we can use train_test_split
X = np.array(X)
Y = np.array(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=23)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled= scaler.fit_transform(X_train)
```

```python
X_test_scaled= scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logr= LogisticRegression(random_state=0).fit(X_train_scaled,y_train)
```

```python
logr.predict(X_train_scaled)
```

```python
logr.score(X_train_scaled,y_train)
```

```python
logr.score(X_test_scaled,y_test)
```
